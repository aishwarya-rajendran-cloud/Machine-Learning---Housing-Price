**FEATURE ENGINEERING** - ***ALGORITHM SELECTION***

```python
#separating good quality wine using conditional statements
#and creating a new coulumn 'gdquality' which has 0's and 1's
wq_data['gdquality'] = [1 if x >= 6 else 0 for x in wq_data['quality']]
```

```python
#Checking rows and column in data
wq_data.dtypes
```

```python
#Dependent and Independent variable split
X = wq_data.drop(['quality','gdquality'],axis=1) #independent
Y = wq_data['gdquality'] #Dependent
```

```python
X.dtypes
```

```python
Y.head() #here the 1's are for quality >=6
```

```python
#Train and Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.25, random_state = 19)
```

```python
Y_test.count()
```

```python
X_train.count()
```

In Supervised learning classification model, List of known algorithms are


1.   Support Vector Machine
2.   Naive Bayes Classifier
1.   Logistic Regression/ Binary Classification
2.   Random Forest
1.   K-Means clustering
2.   KNN algorithm
1.   Decision Tree
2.   Xgboosting algorithm









```python
#Support VeCtor Algorithm
#SVC - Support Vector Classifier
from sklearn.svm import SVC
model1 = SVC()
model1.fit(X_train,Y_train)
pred_y1 = model1.predict(X_test)

#To check the accuracy
from sklearn.metrics import accuracy_score
svm = accuracy_score(Y_test,pred_y1)
```

```python
print(svm)
```

```python
#Naive Bayes Algorithm
from sklearn.naive_bayes import GaussianNB
model2 = GaussianNB()
model2.fit(X_train,Y_train)
pred_y2 = model2.predict(X_test)

#Accuracy Score
from sklearn.metrics import accuracy_score
nb = accuracy_score(Y_test,pred_y2)
```

```python
print(nb)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model3 = LogisticRegression()
model3.fit(X_train,Y_train)
pred_y3 = model3.predict(X_test)

#Accuracy Score
lr = accuracy_score(Y_test,pred_y3)
```

```python
print(lr)
```

```python
#Confussion Matrix for Logistic Regression
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(Y_test,pred_y3)
```

```python
print(conf_mat)
```

```python
#Random Forest algorithm
from sklearn.ensemble import RandomForestClassifier
model4 = RandomForestClassifier()
model4.fit(X_train,Y_train)
pred_y4 = model4.predict(X_test)

#Accuracy score
from sklearn.metrics import accuracy_score
rf = accuracy_score(Y_test,pred_y4)
```

```python
print(rf)
```

```python
#K-means Algorithm
from sklearn.cluster import KMeans
model5 = KMeans()
model5.fit(X_train,Y_train)
pred_y5= model5.predict(X_test)

#accuracy Score
km = accuracy_score(Y_test,pred_y5)
```

```python
print(km)
```

```python
#K Nearest Neighbour Algorithm
from sklearn.neighbors import KNeighborsClassifier
model6 = KNeighborsClassifier()
model6.fit(X_train,Y_train)
pred_y6 = model6.predict(X_test)

#Accuracy scores
knn = accuracy_score(Y_test,pred_y6)
```

```python
print(knn)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
model7 = DecisionTreeClassifier()
model7.fit(X_train,Y_train)
pred_y7 = model7.predict(X_test)

#accuracu score
dt = accuracy_score(Y_test,pred_y7)
```

```python
print(dt)
```

```python
#XGradiant Boosting Algorithm
from xgboost import XGBClassifier
model8 = XGBClassifier()
model8.fit(X_train,Y_train)
pred_y8 = model8.predict(X_test)

#accuracy score
xgb = accuracy_score(Y_test,pred_y8)
```

```python
print(xgb)
```

```python
all_scores = pd.DataFrame({
    'Model': ['Support Vector Classifier','Naive Bayes','Logistic Regression','Random Forest','K Means','KNN','Decision Tree','XGBoosting'],
    'Score': [svm,nb,lr,rf,km,knn,dt,xgb]})
```

```python
print(all_scores)
```

```python
all_scores = all_scores.sort_values(by='Score',ascending=False)
print(all_scores)
```
